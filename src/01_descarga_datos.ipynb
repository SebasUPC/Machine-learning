```python
from pathlib import Path
import kagglehub

DATA_DIR = Path('data')
RAW_DIR = DATA_DIR / 'raw'
RAW_DIR.mkdir(parents=True, exist_ok=True)

def pick_first_csv(path):
    csvs = glob.glob(os.path.join(path, '**', '*.csv'), recursive=True)
    if not csvs:
        raise FileNotFoundError(f'No CSV files found in {path}')
    return csvs[0]

# Dataset de salud
health_path = kagglehub.dataset_download('tarekmasryo/cancer-risk-factors-dataset')
health_csv = pick_first_csv(health_path)
health_df = pd.read_csv(health_csv)
health_out = RAW_DIR / 'health_raw.csv'
health_df.to_csv(health_out, index=False)
print('Saved health dataset:', health_out)
print('Health columns:', health_df.columns.tolist())

# Dataset de inventario (stock)
stock_path = kagglehub.dataset_download('ziya07/high-dimensional-supply-chain-inventory-dataset')
stock_csv = pick_first_csv(stock_path)
stock_df = pd.read_csv(stock_csv)
stock_out = RAW_DIR / 'stock_raw.csv'
stock_df.to_csv(stock_out, index=False)
print('Saved stock dataset:', stock_out)
print('Stock columns:', stock_df.columns.tolist())

health_df.head()
```
